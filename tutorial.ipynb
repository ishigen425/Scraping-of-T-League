```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
player_data = pd.read_csv('./data/player_stats_table_2019.csv')
print(player_data.shape)
for idx, item in player_data.iterrows():
    name = item['player_name']
    split_name = name.split()
    player_data.loc[idx, 'player_name'] = split_name[0][0] + "." + split_name[1]
    print(idx)

player_data = player_data[player_data["match_number"] > 2]
```

```python
get_points_rate = player_data.loc[:, ["player_name","get_points_rate_on_serve","get_points_rate_on_receive", "team_name"]]
get_points_rate["player_name"] = player_name
get_points_rate.set_index("player_name", inplace=True)
get_points_rate
```

```python
# 男子
man_get_points_rate = get_points_rate[(get_points_rate['team_name'] == 'tt-saitama')
     | (get_points_rate['team_name'] == 'kinoshitameister-tokyo')
     | (get_points_rate['team_name'] == 'okayama-rivets')
     | (get_points_rate['team_name'] == 'ryukyu-asteeda')]

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(man_get_points_rate['get_points_rate_on_serve'], man_get_points_rate['get_points_rate_on_receive'],
                    marker="o",alpha=0.5,c='blue')
for k, v in man_get_points_rate.iterrows():
            ax.annotate(v.name, xy=(v['get_points_rate_on_serve'],v['get_points_rate_on_receive']), size=10, alpha=0.5)
plt.title('サーブ・レシーブ別得点率(男子)')
ax.set_xlabel('サーブ時得点率')
ax.set_ylabel('レシーブ時得点率')
fig.show()
plt.savefig("./data/man_get_points_rate")
```

```python
#女子
woman_get_points_rate = get_points_rate[(get_points_rate['team_name'] == 'top-nagoya')
     | (get_points_rate['team_name'] == 'nissay-redelf')
     | (get_points_rate['team_name'] == 'nipponpaint-mallets')
     | (get_points_rate['team_name'] == 'rkinoshitaabyell-kanagawa')]
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(woman_get_points_rate['get_points_rate_on_serve'], woman_get_points_rate['get_points_rate_on_receive'],
                    marker="o",alpha=0.5,c='red')
for k, v in woman_get_points_rate.iterrows():
            ax.annotate(v.name, xy=(v['get_points_rate_on_serve'],v['get_points_rate_on_receive']), size=10, alpha=0.5)
plt.title('サーブ・レシーブ別得点率(女子)')
ax.set_xlabel('サーブ時得点率')
ax.set_ylabel('レシーブ時得点率')
fig.show()
```
